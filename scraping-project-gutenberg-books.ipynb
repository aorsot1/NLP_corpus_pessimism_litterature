# Collecting All Ebook Links

```python
import requests
from bs4 import BeautifulSoup

phil_urls = []
soc_urls = []
anarch_urls = []
racism_urls = []

for index in [1, 26, 51, 76, 101]:
    url = f"https://www.gutenberg.org/ebooks/bookshelf/57?start_index={index}"
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    for link in soup.find_all('a'):
        curr_url = link.get('href')
        if curr_url[:8] == '/ebooks/' and len(curr_url)>8:
            if curr_url[8] in [str(i) for i in list(range(10))]:
                phil_urls.append(curr_url)

for index in [1, 26]:
    url = f"https://www.gutenberg.org/ebooks/bookshelf/134?start_index={index}"
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    for link in soup.find_all('a'):
        curr_url = link.get('href')
        if curr_url[:8] == '/ebooks/' and len(curr_url)>8:
            if curr_url[8] in [str(i) for i in list(range(10))]:
                soc_urls.append(curr_url)
                
for index in [1, 26]:
    url = f"https://www.gutenberg.org/ebooks/bookshelf/134?start_index={index}"
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    for link in soup.find_all('a'):
        curr_url = link.get('href')
        if curr_url[:8] == '/ebooks/' and len(curr_url)>8:
            if curr_url[8] in [str(i) for i in list(range(10))]:
                soc_urls.append(curr_url)
                
url = f"https://www.gutenberg.org/ebooks/bookshelf/7"
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')

for link in soup.find_all('a'):
    curr_url = link.get('href')
    if curr_url[:8] == '/ebooks/' and len(curr_url)>8:
        if curr_url[8] in [str(i) for i in list(range(10))]:
            anarch_urls.append(curr_url)
            
url = f"https://www.gutenberg.org/ebooks/bookshelf/65"
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')

for link in soup.find_all('a'):
    curr_url = link.get('href')
    if curr_url[:8] == '/ebooks/' and len(curr_url)>8:
        if curr_url[8] in [str(i) for i in list(range(10))]:
            racism_urls.append(curr_url)
            
all_urls = {'Philosophy': phil_urls,
            'Sociology': soc_urls,
            'Anarchism': anarch_urls,
            'Racism': racism_urls}
```

# Reading Ebook Content

```python
data = {'Subject':[],
        'Medium':[],
        'Link':[],
        'Text':[],
        'Author':[],
        'Title':[],
        'Date':[]}
```

```python
for subject in ['Philosophy','Sociology','Anarchism','Racism']:
    
    for i in range(len(all_urls[subject])):

        try:

            url = 'https://www.gutenberg.org'+all_urls[subject][i]
            reqs = requests.get(url)
            soup = BeautifulSoup(reqs.text, 'html.parser')

            for link in soup.find_all('a'):
                curr_url = link.get('href')
                if curr_url is None: continue
                if '.txt' in curr_url:
                    text_file = 'https://www.gutenberg.org'+curr_url

            file = urllib.request.urlopen(text_file)
            text = []
            for line in file:
                text.append(line.decode('utf-8'))
            text = ' '.join(text)

            title = text.split('Gutenberg EBook of ')[1].split(',')[0].strip()
            author = text.split('by')[1].split('\r')[0].strip()

            text = text.split('***')[2].strip()
            text = text.replace('\r','')
            text = text.replace('\ufeff','')
            text = text.replace('\n','')

            data['Subject'].append(subject)
            data['Medium'].append('Book')
            data['Link'].append(url)
            data['Text'].append(text)
            data['Author'].append(author)
            data['Title'].append(title)
            data['Date'].append(None)

        except:

            print(f"FOLLOWING URL FAILED: {all_urls[subject][i]}")
```

```python
data = pd.DataFrame(data)
data
```
