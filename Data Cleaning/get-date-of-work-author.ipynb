```python
import wikipedia
import datefinder
import pandas as pd
import numpy as np
```

```python
def get_author_date(name):
    try:
        text = wikipedia.WikipediaPage(name).summary
        matches = datefinder.find_dates(text)
        dates = [date for date in matches]
        birth_date = dates[0]
        return birth_date
    except:
        return np.nan

def get_work_date(title):
    try:
        text = wikipedia.WikipediaPage(title).content
        text = text.split('==')[0]
        matches = datefinder.find_dates(text)
        dates = [date for date in matches]
        publish_date = dates[0]
        return publish_date
    except:
        return np.nan
```
