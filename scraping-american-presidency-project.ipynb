# Collecting All Links

```python
%%cython -a

import numpy as np
from tqdm.notebook import tqdm
import requests
import urllib
from bs4 import BeautifulSoup

def get_content(store_list, url):
    
    try:
        reqs = requests.get(url)
        soup = BeautifulSoup(reqs.text, 'html.parser')

        for link in soup.find_all('a'):
            curr_url = link.get('href')
            if not isinstance(curr_url, str): continue
            if 'documents' in curr_url and curr_url.count('/')==2:
                if 'category-attributes' not in curr_url and 'presidential-documents-archive-guidebook' not in curr_url:
                    store_list.append(curr_url)

    except:
        pass
    
page_len = 50

valid_urls = []

divide_cat = 1
for page in tqdm(range(round(int(np.ceil(78035/page_len))/divide_cat))):
    url = f"https://www.presidency.ucsb.edu/documents/app-categories/presidential?items_per_page={page_len}&page={page}"
    get_content(valid_urls, url)
    
    
def get_info_from_url(relative_url):
    url = 'https://www.presidency.ucsb.edu/'+relative_url
    file = urllib.request.urlopen(url)
    text = []
    for line in file:
        text.append(line.decode('utf-8'))
    text = ' '.join(text)
    relevant_text = ''.join(text.split('<div class="field-docs-content">')[1].split('<p>')[1:]).split('</div')[0]
    for tag in ['<p>','</p>','\n','<b>','</b>','<i>',"</i>"]:
        relevant_text = relevant_text.replace(tag,'')
    relevant_text = relevant_text.replace('--', '-')
    relevant_text = relevant_text.strip()

    title = text.split('<div class="field-ds-doc-title">')[1].split('</div')[0]
    for tag in ['\n','<h1>','</h1>']:
        title = title.replace(tag,'')
    title = title.strip()

    person = text.split('<h3 class="diet-title">')[1].split('</a')[0]
    for tag in ['\n','<h1>','</h1>']:
        person = person.replace(tag,'')
    person = person.strip().split('>')[-1]

    date = text.split('class="date-display-single"')[1].split('</span>')[0].split('>')[-1]
    
    return relevant_text, title, person, date

data = {'URL':[],
        'Person':[],
        'Date':[],
        'Title':[],
        'Text':[]}

for relative_url in tqdm(valid_urls):
    try:
        relevant_text, title, person, date = get_info_from_url(relative_url)
        data['URL'].append('https://www.presidency.ucsb.edu'+relative_url)
        data['Person'].append(person)
        data['Date'].append(date)
        data['Title'].append(title)
        data['Text'].append(relevant_text)
    except:
        pass
```

```python
data = pd.DataFrame(data)
data
```

```python
data.to_csv('app-presidential.csv')
```
